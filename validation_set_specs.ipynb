# Setup

```python
import matplotlib.pyplot as plt
%matplotlib inline

import os
from os import listdir
from os.path import isfile, join
```

## Check out the images

```python
validation_images_folder = os.getcwd() + '/validation_images/'
```

```python
validation_images = [f for f in listdir(validation_images_folder) if isfile(join(validation_images_folder, f))]
print(len(validation_images))
```

```python
states_dict = {}
for filename in validation_images:
    if "no_plate" not in filename:
        filename = filename[:-4]
        state_code = filename[-2:]
        states_dict[filename] = state_code
```

```python
list_state_codes = list(states_dict.values())
print(len(list_state_codes))
print(list(list_state_codes)[0:5])
```

```python
n_unique_state_codes = len(set(list_state_codes))
n_unique_state_codes
```

```python
plt.figure(figsize=(20,10))
plt.hist(list_state_codes, color = 'maroon', edgecolor = 'black', bins = n_unique_state_codes)
plt.title('histogram of state codes on ' + str(len(list_state_codes)) + ' validation images')
plt.show()
```
